```python
# Célula 1 - Importação de bibliotecas
import pandas as pd
import os
from pathlib import Path

# Célula 2 - Listagem de arquivos disponíveis
raw_path = r"C:\Users\x387162\Desktop\Relatorio_Diario\projeto_etl_dados-main\data\raw"
files = os.listdir(raw_path)
xlsx_files = [f for f in files if f.endswith('.xlsx')]

print("Arquivos disponíveis na pasta raw:")
for i, file in enumerate(xlsx_files, 1):
    print(f"{i}. {file}")

# Célula 3 - Seleção do arquivo
selected_index = int(input("\nDigite o número do arquivo que deseja analisar: ")) - 1
selected_file = xlsx_files[selected_index]
file_path = os.path.join(raw_path, selected_file)

print(f"\nArquivo selecionado: {selected_file}")

# Célula 4 - Carregamento e preparação dos dados
df = pd.read_excel(file_path)

# Converter colunas relevantes
df['DataAtendimento'] = pd.to_datetime(df['DataAtendimento'], format='%d/%m/%Y')
df['Mês'] = df['Mês'].str.lower().str.strip()

# Célula 5 - Filtragem dos meses de interesse
meses_alvo = ['mai', 'jun', 'jul', 'ago', 'set', 'out']
df_periodo = df[df['Mês'].isin(meses_alvo)].copy()

print(f"Total de registros no período Maio-Outubro: {len(df_periodo)}")

# Célula 6 - Análise de pacientes que aparecem em todos os meses
pacientes_por_mes = df_periodo.groupby('NomePaciente')['Mês'].nunique()
pacientes_completos = pacientes_por_mes[pacientes_por_mes == len(meses_alvo)].index

print(f"\nPacientes que apareceram em todos os meses entre Maio e Outubro: {len(pacientes_completos)}")

# Célula 7 - Análise com filtro MV = "sim"
df_mv = df_periodo[df_periodo['MV'] == 'sim']
pacientes_mv_por_mes = df_mv.groupby('NomePaciente')['Mês'].nunique()
pacientes_mv_completos = pacientes_mv_por_mes[pacientes_mv_por_mes == len(meses_alvo)].index

print(f"Pacientes MV que apareceram em todos os meses entre Maio e Outubro: {len(pacientes_mv_completos)}")

# Célula 8 - Lista de pacientes MV completos
print("\nLista de pacientes MV que apareceram em todos os meses:")
for paciente in pacientes_mv_completos:
    print(f"- {paciente}")

# Célula 9 - Geração do relatório
output_path = r"C:\Users\x504693\Documents\projetos\projeto_etl_dados\docs\analise_pacientes.txt"

with open(output_path, 'w', encoding='utf-8') as f:
    f.write("RELATÓRIO DE ANÁLISE DE PACIENTES\n")
    f.write("="*50 + "\n\n")
    f.write(f"Arquivo analisado: {selected_file}\n")
    f.write(f"Período analisado: Maio a Outubro de 2025\n\n")
    f.write(f"Total de pacientes no período: {df_periodo['NomePaciente'].nunique()}\n")
    f.write(f"Pacientes que apareceram em todos os meses: {len(pacientes_completos)}\n")
    f.write(f"Pacientes MV que apareceram em todos os meses: {len(pacientes_mv_completos)}\n\n")
    f.write("Pacientes MV com atendimentos em todos os meses:\n")
    for paciente in pacientes_mv_completos:
        f.write(f"- {paciente}\n")

print(f"\nRelatório salvo em: {output_path}")
```
